```python
# Convertir la columna InvoiceDate a tipo datetime
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"], format="%m/%d/%Y %H:%M")
```

```python
# Resumen estadIstico de variables numericas
summary_stats = df[["Quantity", "UnitPrice"]].describe()
```

```python
# Distribución de Quantity y UnitPrice
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(df["Quantity"], bins=50, kde=True, ax=axes[0])
axes[0].set_title("Distribución de Cantidad Comprada")
axes[0].set_xlabel("Cantidad")

sns.histplot(df["UnitPrice"], bins=50, kde=True, ax=axes[1])
axes[1].set_title("Distribución de Precio Unitario")
axes[1].set_xlabel("Precio Unitario")

plt.tight_layout()
plt.show()

summary_stats
```
